```python
#### task1: Merging excel files:
```

```python
import pandas as pd
import os

df = pd.read_csv('./Sales_Data/Sales_April_2019.csv')

files=[file for file in os.listdir('./Sales_Data')]
all_months_data = pd.DataFrame()

for file in files:
  df = pd.read_csv("./Sales_Data/"+file)
  all_months_data = pd.concat([all_months_data,df])

all_months_data.to_csv("all_data.csv", index=False)
all_months_data.head()
```

```python
#### Read data frame
```

```python
all_data = pd.read_csv("all_data.csv")
all_data.head()
```

```python
####Clean up the Data
```

```python
#### Drop rows of NAN
```

```python
nan_df = all_data[all_data.isna().any(axis=1)]
nan_df.head()
all_data = all_data.dropna(how='all')
all_data.head()
```

```python
#### Find 'OR' and delete it
```

```python
all_data = all_data[all_data['Order Date'].str[0:2] != 'Or']
all_data.head()
```

```python
#### Convert Columns to the Correct Type
```

```python
all_data['Quantity Ordered'] = pd.to_numeric(all_data['Quantity Ordered'])  # Make int
all_data['Price Each'] = pd.to_numeric(all_data['Price Each'])      # Make float

all_data.head()
```

```python
# Augment data with additional columns
```

```python
# Task2: Add Month column
```

```python
all_data['Month']=all_data['Order Date'].str[0:2]
all_data['Month']=all_data['Month'].astype('int32')
all_data.head()
```

```python
#### Add a sales column:
```

```python
all_data['Sales']=all_data['Quantity Ordered'] * all_data['Price Each']
all_data.head()
```

#### Question1: Best Month for Sales & How much was earned that month?

```python
results = all_data.groupby('Month').sum()
```

```python
import matplotlib.pyplot as plt

months = range(1,13)

plt.bar(months, results['Sales'])
plt.xticks(months)
plt.ylabel('Sales in USD ($)')
plt.xlabel('Month Number')
plt.show()
```

Question 2: What city had the highest number of sales?

```python
#### Add a City Column
# let's use .apply()
#### Add a City Column
# let's use .apply()

def get_city(address):
  return address.split(',')[1]

def get_state(address):
  return address.split(',')[2]

def get_zipcode(address):
  return address.split(',')[2].split(' ')[2]

#all_data['City']=all_data['Purchase Address'].apply(lambda x: x.split(',')[1])  # is not the most efficent but works for daily ds jobs
all_data['City']= all_data['Purchase Address'].apply(lambda x: get_city(x))      # More effecient
all_data['State']= all_data['Purchase Address'].apply(lambda x: get_state(x))
all_data['zipcode']= all_data['Purchase Address'].apply(lambda x: get_zipcode(x))



```

```python
results = all_data.groupby('City').sum()
selected_columns = ['Quantity Ordered', 'Price Each', 'Month', 'Sales', 'City']  # Replace with actual column names

```

```python
import matplotlib.pyplot as plt

cities = [city for city, df in all_data.groupby('City')]

plt.bar(cities, results['Sales'])
plt.xticks(cities, rotation='vertical', size=8)
plt.xlabel('City')
plt.ylabel('Sales in the City in USD ($)')
plt.show()
```
